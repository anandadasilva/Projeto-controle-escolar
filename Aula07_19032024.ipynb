```python
def cadastrar_aluno():

 while True:
     matricula = input("digite a matricula: ")
     if not matricula in dic_aluno.keys():
          if len(matricula)==0:
              break
          nome = input("digite nome: ")
          email = input("digite email: ")
          dic_aluno[matricula]=[nome,email]
          dic_banco['aluno']=dic_aluno
     else:
           print("matricula ja existente")
```

```python
def cadastrar_curso():
   while True:
     codigo= input("Informe o código: ")
     if not codigo in dic_curso.keys():
         if len(codigo)==0:
            break
         nome= input("Informe o curso: ")
         dic_curso[codigo]=nome
         dic_banco['curso']=dic_curso
     else:
         print("curso existente")



```

```python
def matricular_aluno():
   imprimeCurso()
   print("Qual curso você deseja se matricular?")
   while True:
     codigo= input("Informe o codigo do curso: ")
     if len(codigo)==0:
        break
     if not codigo in dic_curso.keys():
        print("não existe esse curso")
     else:
        imprimeAluno()
        while True:
          matricula = input("Informe a matrícula do aluno: ")
          if len(matricula)==0:
             break
          if not matricula in dic_alunocurso.keys():
             print("Não existe esse aluno")
          dic_alunocurso[codigo + matricula]=[time.ctime()]
          dic_banco['alunocurso']=dic_alunocurso

```

```python
while True:
  print("sistema de controle escolar")
  print("1-cadastrar aluno")
  print("2-cadastrar curso")
  print("3-matricular aluno")
  print("4-sair")
  opcao=input("escolha opcão")
  if int(opcao)==1:
    cadastrar_aluno()
  elif int(opcao)==2:
    cadastrar_curso()
  elif int(opcao)==3:
    matricular_aluno()
  elif int(opcao)==4:
    break
  else:
    print("opcao invalida")

```
